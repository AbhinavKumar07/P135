```python
import pandas as pd
import matplotlib.pyplot as plt

gravity_filtered_stars = pd.read_csv('https://drive.google.com/uc?id=1TI_uvkBcHxvpa7IDEsBfTdpiacgEA3s9')

gravity_filtered_stars.head(21)
```

```python
star_name_list = sorted(gravity_filtered_stars["Star_name"].tolist())
star_distance_list = sorted(gravity_filtered_stars["Distance"].tolist())
star_mass_list = sorted(gravity_filtered_stars["Mass"].tolist())
star_radius_list = sorted(gravity_filtered_stars["Radius"].tolist())
star_gravity_list = sorted(gravity_filtered_stars["Gravity"].tolist())

plt.figure(figsize = (10,5))
plt.title("Star Gravity")
plt.bar(star_name_list[0:9] , star_gravity_list[0:9])
```

```python
plt.figure(figsize = (11,5))
plt.title("Star Gravity 2")
plt.bar(star_name_list[10:20] , star_gravity_list[10:20])
```
